    Training, testing, and decision flow:

    1) Train each model and generate probabilities in the validation/test;

    2) Scan thresholds from 0 to 1 and calculate:
    Cost = CFP * FP + CFN * FN
    Where we well take as:
    -> CFP = 1
    -> CFN = 5

    3) Take the threshold that minimizes cost for each model;

    4) Compare the minimum cost + auxiliary metrics (PR AUC, ROC AUC).

    Obs.: Keep in mind that this data comes from superficial business rules. We assume it's better to classify a paying customer as delinquent than to incur losses due to delinquency.
    FP (False Positive): You predicted 1 (default), but the person was 0 (good)
    → You denied a loan that could have been repaid (lost profit/opportunity, perhaps customer dissatisfaction).

    FN (False Negative): You predicted 0 (good), but the person was 1 (default)
    → You approved someone who will default (lost principal + collection costs + provision + capital + regulatory risk, etc.).
  </div>
</div>

```r
rec_lr <- recipe(default_2y ~ ., data = train) %>%
  step_dummy(all_nominal_predictors()) %>%
  step_zv(all_predictors()) %>%
  step_normalize(all_numeric_predictors())
```

```r
# Cost function and the search for the best threshold
# Cost function given a threshold
cost_at_t <- function(truth, p1, t, c_fp = 1, c_fn = 5) {
  pred <- factor(ifelse(p1 >= t, "1", "0"), levels = c("0", "1"))
  truth <- factor(truth, levels = c("0", "1"))
  
  fp <- sum(pred == "1" & truth == "0")
  fn <- sum(pred == "0" & truth == "1")
  
  c_fp * fp + c_fn * fn
}

# Sweep thresholds
ts <- seq(0, 1, by = 0.01)
costs <- sapply(ts, function(t) cost_at_t(pred_test$default_2y, pred_test$.pred_1, t, c_fp = 1, c_fn = 5))

best_t <- ts[which.min(costs)]
best_cost <- min(costs)

best_t
best_cost
```

    Given that the probability is p
    My chance of default being 0 is (1-p) and my chance of being 1 is p
    So my expected cost if I predict 1 is CFP*(1-p)
    And my expected cost if I predict 0 is CFN*p

    The lowest expected cost, according to my initial hypothesis, is if I incorrectly classify as 1, when it should be 0. So the lowest cost is
    CFP(1-p)
    Which in turn is lower than my hypothesis of classifying as compliant and being in default. So:
    CFP(1-p) < CFN*p

    Which arrives at the equation through algebra:
    p > (CFP)/(CFP + CFN)

    Therefore, my theoretical probability that yields the lowest possible theoretical cost is:
    t > (CFP)/(CFP + CFN)
    I compare the two expected costs and choose the class with the lower cost for that price.

    Substituting the values ​​based on the hypothesis that CFP = 1 and CFN = 5, we arrive at the value of:

    t = 1/6

    This is the theoretical value that will form the basis for the XGBoost hyperparameter tuning iterations in the search for parameters that yield the lowest cost.

```r

# Define the hyperparameters to be applied to the tuning

mod_xgb_tune <- boost_tree(
  trees = 500,
  tree_depth = tune(),
  min_n = tune(),
  learn_rate = tune()
) %>%
  set_engine("xgboost") %>%
  set_mode("classification")

wf_xgb_tune <- workflow() %>%
  add_recipe(rec_xgb) %>%
  add_model(mod_xgb_tune)
```

```r
# Create a test grid

grid_xgb <- grid_regular(
  tree_depth(range = c(3L, 9L)),
  min_n(range = c(1L, 30L)),
  learn_rate(range = c(-2, -1)),
  levels = c(3, 3, 2)
)

#grid_xgb
```

```r
# to run the tuning with CV
res_xgb_tune <- tune_grid(
  wf_xgb_tune,
  resamples = folds,
  grid = grid_xgb,
  control = control_grid(save_pred = TRUE, verbose = TRUE)
)
```

```r
# Choose the best one based on COST (simple)
# cost function (assumes default_2y levels are "0","1")
cost_at_t <- function(truth, p1, t, c_fp = 1, c_fn = 5) {
  truth <- factor(truth, levels = c("0","1"))
  pred  <- factor(ifelse(p1 >= t, "1", "0"), levels = c("0","1"))
  fp <- sum(pred == "1" & truth == "0")
  fn <- sum(pred == "0" & truth == "1")
  c_fp * fp + c_fn * fn
}

t_fixed <- 1/(1+5)

preds <- collect_predictions(res_xgb_tune)

cost_by_combo <- preds %>%
  group_by(.config, tree_depth, min_n, learn_rate) %>%
  summarise(cost = cost_at_t(default_2y, .pred_1, t = t_fixed, c_fp = 1, c_fn = 5),
            .groups = "drop") %>%
  arrange(cost)

head(cost_by_combo, 10)
```
