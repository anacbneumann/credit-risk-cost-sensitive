  <div style="font-size: 13px; line-height: 1.55;">
    <!-- Write your comment here -->

    0.41% of the dataset contains duplicate rows (excluding the ID). This may indeed represent coincidences of age x income x dependents x debt rate, etc. However, considering that the probability of repeating this data combination is very low for the same local database, we will consider it as duplicate data input errors. The data may have been entered twice by the responsible party while maintaining different IDs. We also need to consider that in the future, for the application of the models, we will remove the ID column, and duplicates can introduce false biases to the application.
    
    Furthermore, we can consider the possibility of one of the duplicate rows leaking into the validation: if a random train/valid split is performed and the same duplicate row exists, one copy may appear in the training and another in the test → the metric becomes artificially better (it seems that the model generalizes, but it only "recognized" the same record).
      
  </div>
</div>

```r
# Defining IQR and Boxplot functions to apply to each column.

# IQR report 
iqr_report <- function(df, col, id_col = "id", show = 20) {
  x <- df[[col]]

  q1  <- as.numeric(quantile(x, 0.25, na.rm = TRUE))
  q3  <- as.numeric(quantile(x, 0.75, na.rm = TRUE))
  iqr <- q3 - q1

  lower <- q1 - 1.5 * iqr
  upper <- q3 + 1.5 * iqr

  out_idx <- which(!is.na(x) & (x < lower | x > upper))

  cat(sprintf("Column: %s\n", col))
  cat(sprintf("Q1: %.4f | Median: %.4f | Q3: %.4f | IQR: %.4f\n",
              q1, median(x, na.rm = TRUE), q3, iqr))
  cat(sprintf("Fences (1.5*IQR): [%.4f, %.4f]\n", lower, upper))
  cat(sprintf("Outliers: %d (%.4f%%)\n",
              length(out_idx), 100 * length(out_idx) / sum(!is.na(x))))

  # return a list
  out_vals <- x[out_idx]
  out_ids  <- if (id_col %in% names(df)) df[[id_col]][out_idx] else out_idx

  invisible(list(
    col = col,
    q1 = q1, q3 = q3, iqr = iqr,
    lower = lower, upper = upper,
    outlier_n = length(out_idx),
    outlier_pct = 100 * length(out_idx) / sum(!is.na(x)),
    outlier_idx = out_idx,
    outlier_ids = out_ids,
    outlier_values = out_vals,
    preview = head(sort(out_vals), show)
  ))
}

# BoxPlot
plot_box <- function(df, col, color = "#5B4661", log1p = FALSE, outline = TRUE) {
  x <- df[[col]]
  x <- x[!is.na(x)]
  if (log1p) x <- log1p(x)
  op <- par(bg = "white", mar = c(4, 4, 3, 1) + 0.1)
  on.exit(par(op))

  boxplot(
    x,
    horizontal = TRUE,
    col = adjustcolor(color, alpha.f = 0.25),
    border = color,
    medcol = color,
    whiskcol = color,
    staplecol = color,
    outcol = color,
    outpch = 16,
    outline = outline,  
    main = paste0("Boxplot - ", col, if (log1p) " (log1p)" else ""),
    xlab = if (log1p) paste0("log1p(", col, ")") else col
  )

  invisible(TRUE)
}
```
