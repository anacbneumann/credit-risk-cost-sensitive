  <div style="font-size: 13px; line-height: 1.55;">
    <!-- Write your comment here -->

    0.41% of the dataset contains duplicate rows (excluding the ID). This may indeed represent coincidences of age x income x dependents x debt rate, etc. However, considering that the probability of repeating this data combination is very low for the same local database, we will consider it as duplicate data input errors. The data may have been entered twice by the responsible party while maintaining different IDs. We also need to consider that in the future, for the application of the models, we will remove the ID column, and duplicates can introduce false biases to the application.
    
    Furthermore, we can consider the possibility of one of the duplicate rows leaking into the validation: if a random train/valid split is performed and the same duplicate row exists, one copy may appear in the training and another in the test → the metric becomes artificially better (it seems that the model generalizes, but it only "recognized" the same record).
      
  </div>
</div>

```r
# Missingness vs target (default_2y)
#MCAR/MAR

target_col <- "default_2y"
target     <- df[[target_col]]
target_fac <- factor(target, levels = c(0, 1))

miss_vs_target <- do.call(rbind, lapply(miss_cols, function(col) {
  miss <- is.na(df[[col]])
  if (sum(miss) == 0) return(NULL)

  rate_miss   <- mean(target[miss] == 1, na.rm = TRUE)
  rate_nomiss <- mean(target[!miss] == 1, na.rm = TRUE)

  pval <- NA
  tab <- table(miss, target_fac)
  if (all(dim(tab) == c(2, 2)) && all(tab > 0)) {
    pval <- suppressWarnings(chisq.test(tab)$p.value)
  }

  data.frame(
    column = col,
    n_missing = sum(miss),
    pct_missing = round(100 * mean(miss), 2),
    default_rate_missing = round(100 * rate_miss, 2),
    default_rate_not_missing = round(100 * rate_nomiss, 2),
    p_value = pval
  )
}))

miss_vs_target <- miss_vs_target[order(-miss_vs_target$pct_missing), ]
miss_vs_target
```
