```r
# Checking duplicate rows
cat('\nCounting how many IDs are repeated:\n')
dup_id  <- sum(duplicated(cs_train$id))
dup_id

cat('\nCounting how many complete rows are repeated (all columns are the same):\n')
dup_row <- sum(duplicated(cs_train))
dup_row
```

```r
# Duplicated rows ignoring the ID column
cat('\nCounting how many complete rows (ignoring ID) are repeated (all columns are the same):\n')

dup_mask_no_id <- duplicated(cs_train[, setdiff(names(cs_train), "id")])

dup_n <- sum(dup_mask_no_id)
dup_pct <- 100 * dup_n / nrow(cs_train)

cat(sprintf("Repeated rows (ignoring ID): %d (%.2f%% of dataset)\n", dup_n, dup_pct))

dup_rows_no_id <- cs_train[
  dup_mask_no_id | duplicated(cs_train[, setdiff(names(cs_train), "id")], fromLast = TRUE),
]

dup_rows_no_id_sorted <- dup_rows_no_id[order(-dup_rows_no_id$age_years), ]
head(dup_rows_no_id_sorted, 20)
```

  <div style="font-size: 13px; line-height: 1.55;">
    <!-- Write your comment here -->

    0.41% of the dataset contains duplicate rows (excluding the ID). This may indeed represent coincidences of age x income x dependents x debt rate, etc. However, considering that the probability of repeating this data combination is very low for the same local database, we will consider it as duplicate data input errors. The data may have been entered twice by the responsible party while maintaining different IDs. We also need to consider that in the future, for the application of the models, we will remove the ID column, and duplicates can introduce false biases to the application.
    
    Furthermore, we can consider the possibility of one of the duplicate rows leaking into the validation: if a random train/valid split is performed and the same duplicate row exists, one copy may appear in the training and another in the test → the metric becomes artificially better (it seems that the model generalizes, but it only "recognized" the same record).
      
  </div>
</div>

```r
# Removing duplicated rows - keeping 1 of the duplicated rows
cs_train_dedup <- cs_train[!duplicated(cs_train[, setdiff(names(cs_train), "id")]), ]
```

```r
# Checking again the duplicated rows
cat('\nCounting how many complete rows (ignoring ID) are repeated (all columns are the same):\n')

dup_mask_no_id <- duplicated(cs_train_dedup[, setdiff(names(cs_train_dedup), "id")])

dup_n <- sum(dup_mask_no_id)
dup_pct <- 100 * dup_n / nrow(cs_train_dedup)

cat(sprintf("Repeated rows (ignoring ID): %d (%.2f%% of dataset)\n", dup_n, dup_pct))

dup_rows_no_id <- cs_train_dedup[
  dup_mask_no_id | duplicated(cs_train_dedup[, setdiff(names(cs_train_dedup), "id")], fromLast = TRUE),
]

dup_rows_no_id_sorted <- dup_rows_no_id[order(-dup_rows_no_id$age_years), ]
head(dup_rows_no_id_sorted, 20)
```
